# **Data Quality Check**

Handle Missing Values

```python
data.isnull().sum()
#Findinga the null values in each columns
#Only age column has null values
```

```python
data['age'] = data['age'].fillna(data['age'].mean())
#filling "age" missing values with their column mean values.
```

```python
plt.figure(figsize=(5,3))
data['age'].plot(kind='kde')
plt.show()
# Used mean to fill null values in age columns so the distribution will be normal
```

```python
data.dtypes
#Checking data type of each columns
```

```python
data['age'] = data['age'].astype(int)
#Changing the data type of age from float to int
```

```python
data['purchase_date'] = pd.to_datetime(data['purchase_date'])
#Changing the data type of purchase_date from object to datetime
```

```python
# What is the min and max of purchase_date? Does it match Jan 2023 – Dec 2025?
data['purchase_date'].min()
data['purchase_date'].max()
# Yes the min and max of purchase_data matches
# For min Purchase_date = January 2023
# For max Purchase_date = December 2025
```

```python
# How many unique customers, cities, categories exist?
data.groupby('customer_id')['customer_id'].nunique()
data.groupby('city')['city'].nunique() #15 Uniques values present in dataset
data.groupby('product_category')['product_category'].nunique() #7 Uniques values present in dataset
```

# **Feature Engineering**

```python
#Extracting month , year and date from purchase_date
import datetime
data['Month_Name'] = data['purchase_date'].dt.strftime('%B')
data['Year'] = data['purchase_date'].dt.year
data['Date'] = data['purchase_date'].dt.day
data['Day'] = data['purchase_date'].dt.day_name()
```

```python
data.head(5)
```

```python
#Categorised age column as per their age
data['Category'] = data['age'].apply(lambda x: 'Adult' if x <= 35 else 'Senior Citizen')
```

```python
data.sample(5)
```

```python
#What day of week did each transaction happen?
data.groupby('Day')['Day'].nunique()
#Everyday of week transaction happend.
```

# Bivariate Analysis

```python
#city and total_amount
plt.figure(figsize=(10,4))
sns.barplot(x='city', y='total_amount', data=data,hue='city')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#lucknow gave use maximum revenue
#surat gave use minimum revenue
```

```python
# Group by city and find the sum of 'total_amount'
city_total_amount = data.groupby('city')['total_amount'].sum()

max_city_amount = city_total_amount.idxmin()
max_amount = city_total_amount.min()

#max_city_amount = city_total_amount.idxmax()
#max_amount = city_total_amount.max()

print(f"The city with the maximum total amount is '{max_city_amount}' with a total of {max_amount:.2f}.")
```

```python
#product_category and total_amount
plt.figure(figsize=(10,3))
sns.barplot(x='product_category', y='total_amount', data=data,hue='product_category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#Electronic made maximum revenue and Books made minimum revenue
```

```python
#Customer and total_amount
plt.figure(figsize=(10,3))
sns.boxplot(x='customer_segment', y='total_amount', data=data,hue='customer_segment')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#payment and total_amount
plt.figure(figsize=(10,3))
sns.barplot(x='payment_method', y='total_amount', data=data,hue='payment_method')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#customer prefer UPI and least use Wallet
```

```python
#city and product_category
pivot = data.pivot_table(index='city', columns='product_category', values='total_amount', aggfunc='sum')
pivot
```

```python
plt.figure(figsize=(15, 5))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="viridis", linewidths=.8)
plt.title('Total Amount spend by cities on Product Category')
plt.xlabel('Product Category')
plt.ylabel('City')
#plt.xticks(rotation=45, ha='right')
#plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
# maximum cities spend on electronics
```

```python
#Customer_segment and product_category
product_segment_sales = data.groupby(['product_category', 'customer_segment'])['total_amount'].sum().unstack()
product_segment_sales
```

```python
product_segment_sales.plot(kind='bar', figsize=(5,5))
plt.title('Total Amount by Customer Segment on product ')
plt.xlabel('Product Category')
plt.ylabel('Total Amount')
plt.legend(title='Customer Segment')
plt.tight_layout()
plt.show()
```

```python
#month and total_amount
monthly_sales = data.groupby('Month_Name')['total_amount'].sum()
monthly_sales
```

```python
plt.figure(figsize=(5, 8))
plt.pie(monthly_sales, labels=monthly_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Sales Distribution by Month')
plt.ylabel('')
plt.tight_layout()
plt.show()
#our maximum revenue is occured in month may
# least revenue is occuered in month feb
```

```python
#total-amount and product segment
revenue = data.groupby('customer_segment')['total_amount'].sum()
revenue
```

```python
plt.figure(figsize=(5,4))
plt.pie(revenue,labels = revenue.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Revenue by customer segment ')
plt.ylabel('')
plt.tight_layout()
plt.show()
#our maximum revenue is made my premium customers which is 50.3% of total revenue
```
